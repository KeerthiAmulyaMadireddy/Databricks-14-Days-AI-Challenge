```python
%sql
USE CATALOG ecommerce;
-- checking daily events in gold
CREATE OR REPLACE VIEW gold.daily_events AS 
SELECT
  DATE(event_time) AS event_date,
  event_type,
  brand,
  category_code,
  category_id,
  COUNT(*) AS events,
  COUNT(DISTINCT product_id) AS unique_products
FROM silver.events
GROUP BY
  DATE(event_time), event_type, brand, category_code, category_id;
```

```python
%sql
SELECT * FROM gold.daily_events ORDER BY event_date DESC LIMIT 20;
```

### Revenue trend alternative: “Engagement trend” (daily)

```python
%sql
USE CATALOG ecommerce;

SELECT
  DATE(event_time) AS event_date,
  SUM(CASE WHEN event_type = 'view' THEN 1 ELSE 0 END) AS views,
  SUM(CASE WHEN event_type IN ('cart','add_to_cart') THEN 1 ELSE 0 END) AS carts,
  SUM(CASE WHEN event_type IN ('purchase','order','checkout_complete') THEN 1 ELSE 0 END) AS purchases
FROM silver.events
GROUP BY DATE(event_time)
ORDER BY event_date;
```

### Funnel (view → cart → purchase) with conversion rates

```python
%sql
USE CATALOG ecommerce;

WITH daily AS (
  SELECT
    DATE(event_time) AS event_date,
    SUM(CASE WHEN event_type = 'view' THEN 1 ELSE 0 END) AS views,
    SUM(CASE WHEN event_type IN ('cart','add_to_cart') THEN 1 ELSE 0 END) AS carts,
    SUM(CASE WHEN event_type IN ('purchase','order','checkout_complete') THEN 1 ELSE 0 END) AS purchases
  FROM silver.events
  GROUP BY DATE(event_time)
)
SELECT
  event_date,
  views,
  carts,
  purchases,
  carts / NULLIF(views, 0)      AS view_to_cart_rate,
  purchases / NULLIF(carts, 0)  AS cart_to_purchase_rate,
  purchases / NULLIF(views, 0)  AS view_to_purchase_rate
FROM daily
ORDER BY event_date;
```

### Top products (by views/carts/purchases)

```python
%sql
USE CATALOG ecommerce;

SELECT
  product_id,
  MAX(brand) AS brand,
  MAX(category_code) AS category_code,
  SUM(CASE WHEN event_type = 'view' THEN 1 ELSE 0 END) AS views,
  SUM(CASE WHEN event_type IN ('cart','add_to_cart') THEN 1 ELSE 0 END) AS carts,
  SUM(CASE WHEN event_type IN ('purchase','order','checkout_complete') THEN 1 ELSE 0 END) AS purchases
FROM silver.events
GROUP BY product_id
ORDER BY views DESC
LIMIT 20;
```

## Top brands trend (daily)

```python
%sql
USE CATALOG ecommerce;

SELECT
  DATE(event_time) AS event_date,
  brand,
  COUNT(*) AS events
FROM silver.events
WHERE brand IS NOT NULL
GROUP BY DATE(event_time), brand
ORDER BY event_date, events DESC;
```

```python
%sql
USE CATALOG ecommerce;

SELECT
  CASE
    WHEN cnt >= 100 THEN 'VIP'
    WHEN cnt >= 50  THEN 'Loyal'
    ELSE 'Regular'
  END AS tier,
  COUNT(*) AS products
FROM (
  SELECT
    product_id,
    COUNT(*) AS cnt
  FROM silver.events
  GROUP BY product_id
)
GROUP BY tier;
```
