```python
%sql
CREATE CATALOG IF NOT EXISTS ecommerce;

USE CATALOG ecommerce;

--Create medallion schemas
CREATE SCHEMA IF NOT EXISTS bronze;
CREATE SCHEMA IF NOT EXISTS silver;
CREATE SCHEMA IF NOT EXISTS gold;
```

```python
base_path   = "/Volumes/workspace/ecommerce/ecommerce_data"

bronze_path = f"{base_path}/bronze_events"
silver_path = f"{base_path}/silver_events"
gold_path   = f"{base_path}/gold_products"

# Register Bronze
bronze_df = spark.read.format("delta").load(bronze_path)
bronze_df.write.mode("overwrite").saveAsTable("ecommerce.bronze.events")

# Register Silver
silver_df = spark.read.format("delta").load(silver_path)
silver_df.write.mode("overwrite").saveAsTable("ecommerce.silver.events")

# Register Gold
gold_df = spark.read.format("delta").load(gold_path)
gold_df.write.mode("overwrite").saveAsTable("ecommerce.gold.products")
```

```python
%sql
USE CATALOG ecommerce;

CREATE OR REPLACE VIEW gold.top_products AS
SELECT
    product_id,
    brand,
    revenue,
    conversion_rate
FROM gold.products
WHERE purchases > 10
ORDER BY revenue DESC
LIMIT 100;
```
